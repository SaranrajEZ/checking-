```python
import pandas as pd
import datetime as dt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt

# Define the start and end dates
start = dt.datetime(2021, 6, 1)
end = dt.datetime(2022, 1, 5)
symbol = 'AXISBANK.NS'

# Fetch stock data
stk_data = yf.download(symbol, start=start, end=end)
stk_data = stk_data[['Open', 'High', 'Low', 'Close']]

# Normalize data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(stk_data[['Close']])
data_normalized = pd.DataFrame(data_normalized, columns=["Close"])
```

```python

# Implement Simple Exponential Smoothing (SES)
alpha = 0.8  # Smoothing parameter (adjust as needed)
ses_forecast = []
previous_forecast = data_normalized["Close"].iloc[0]  # Initialize with the first "Close" value

for i in range(len(data_normalized)):
    value = data_normalized["Close"].iloc[i]
    forecast = alpha * value + (1 - alpha) * previous_forecast
    ses_forecast.append(forecast)
    previous_forecast = forecast

# Inverse transform forecast to original scale
forecasted_prices = scaler.inverse_transform(pd.DataFrame(ses_forecast, columns=["Close"]))

# Calculate performance metrics
actual_prices = stk_data["Close"].values[-len(ses_forecast):]
rmse = mean_squared_error(actual_prices, forecasted_prices, squared=False)
mape = mean_absolute_percentage_error(actual_prices, forecasted_prices)

# Display performance metrics
performance = {
    "Model": ["SES"],
    "RMSE": [rmse],
    "MAPE": [mape]
}
perf = pd.DataFrame(performance)
print(perf)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(stk_data.index[-len(ses_forecast):], actual_prices, color='blue', label="Actual")
plt.plot(stk_data.index[-len(ses_forecast):], forecasted_prices, color='green', label="Forecasted (SES)")
plt.title("AXISBANK-Close-SES Forecast")
plt.xlabel("Date")
plt.ylabel("Prices")
plt.legend()
plt.show()
```
