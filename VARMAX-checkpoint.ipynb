```python
import pandas as pd
import datetime as dt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.api import VARMAX
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import warnings

# Define the start and end dates
start = dt.datetime(2021, 6, 1)
end = dt.datetime(2022, 1, 5)
symbol = 'AXISBANK.NS'

# Fetch stock data
stk_data = yf.download(symbol, start=start, end=end)
stk_data = stk_data[['Open', 'High', 'Low', 'Close']]

# Normalize the data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(stk_data)
data_normalized = pd.DataFrame(data_normalized, columns=["Open", "High", "Low", "Close"])

# Suppress warnings
warnings.filterwarnings("ignore")

# Print the selected columns
listt = ["Open", "High", "Low", "Close"]
print(listt)
datasetTwo = data_normalized[listt]

# Split the dataset into training and testing sets
test_obs = 28
train = datasetTwo[:-test_obs]
test = datasetTwo[-test_obs:]
```

```python
warnings.filterwarnings("ignore")
```

```python
rmse_list = []
mape_list = []
lag_list = []

# Iterate through combinations of p and q
for p in range(1, 6):
    for q in range(1, 4):
        try:
            model = VARMAX(train, order=(p, q))
            result = model.fit(maxiter=1000, disp=False)

            print('Order = (p={}, q={})'.format(p, q))
            print('AIC: ', result.aic)
            print('BIC: ', result.bic)
            print()

            pred = result.forecast(steps=28)

            rmse = round(mean_squared_error(test, pred, squared=False))
            mape = mean_absolute_percentage_error(test, pred)

            rmse_list.append(rmse)
            mape_list.append(mape)
            lag_list.append((p, q))
        except Exception as e:
            print(f"Error fitting model for order (p={p}, q={q}): {e}")

if not rmse_list:
    print("No suitable models found.")
else:
    min_rmse_index = rmse_list.index(min(rmse_list))
    best_lag = lag_list[min_rmse_index]
    best_rmse = rmse_list[min_rmse_index]
    best_mape = mape_list[min_rmse_index]

    print('Best Model (based on RMSE):')
    print('Order (p, q):', best_lag)
    print('RMSE:', best_rmse)
    print('MAPE:', best_mape)

    best_model = VARMAX(train, order=best_lag)
    best_result = best_model.fit(maxiter=1000, disp=False)

    best_pred = best_result.forecast(steps=28)

    preds = pd.DataFrame(best_pred, columns=listt)
    preds.to_csv("varmax_forecasted_{}.csv".format(test_obs))

    performance = pd.DataFrame({
        'Model': [listt],
        'RMSE': [best_rmse],
        'MAPE': [best_mape],
        'Lag': [best_lag],
        'Test': [test_obs]
    })

    print(performance)
```
